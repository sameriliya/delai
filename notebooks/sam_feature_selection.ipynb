```python
column_subset = ["FlightDate",
                 "Year", 
                 'Quarter', 
                 'Month', 
                 'DayofMonth', 
                 'DayOfWeek',
                 "Airline",
                 "Operating_Airline",
                 "Marketing_Airline_Network",
                 "Flight_Number_Marketing_Airline",
                 "Origin",
                 "Dest",
                 "CRSDepTime",
                 "OriginAirportID", 
                 "OriginCityName", 
                 "OriginStateName", 
                 "DestAirportID",
                 "DestCityName", 
                 "DestStateName", 
                 "CRSArrTime",
                 "Distance",
                 #"Flight time", we're going to derive this later! 
                 "ArrDelayMinutes",
                 "Cancelled",
                 "Diverted"] 
```

```python
print(csv_files)
```

```python
csv_2019 = pd.read_csv('../raw_data/Combined_Flights_2019.csv')

csv_2019 = csv_2019[column_subset]
```

```python
csv_2018 = pd.read_csv('../raw_data/Combined_Flights_2018.csv')

csv_2018 = csv_2018[column_subset]
```

```python
csv_2020 = pd.read_csv('../raw_data/Combined_Flights_2020.csv')

csv_2020 = csv_2020[column_subset]
```

```python
csv_2021 = pd.read_csv('../raw_data/Combined_Flights_2021.csv')

csv_2021 = csv_2021[column_subset]
```

```python
csv_2022 = pd.read_csv('../raw_data/Combined_Flights_2022.csv')

csv_2022 = csv_2022[column_subset]
```

```python
csv_2022.head()
```

```python
combined = pd.concat([csv_2018,csv_2019, csv_2020, csv_2021, csv_2022], ignore_index=True)
```

```python
combined.to_csv('../raw_data/combined_subset.csv')
```

```python
percent_missing = combined.isnull().sum() * 100 / len(combined)
percent_missing
```

## Train Test Split

```python
combined.head()
```

```python
X = combined.drop(columns = ['ArrDelayMinutes','Cancelled','Diverted'])
X.head()
```

```python
y = combined[['ArrDelayMinutes','Cancelled','Diverted']]
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```
