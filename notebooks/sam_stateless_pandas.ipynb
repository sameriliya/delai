```python
def return_dummies_df(df):
    cat_data = df[['Origin','Dest','Marketing_Airline_Network']]
    dest_dummies = pd.get_dummies(cat_data, prefix=['o','d','a'], ).astype(int)
    return df.merge(dest_dummies, left_index=True, right_index=True)
```

# Repeating this for Train, Val, Test

```python
from google.cloud import bigquery

def save_df_chunk_to_bq(df, table_name, is_first):
    client = bigquery.Client()
    table_id = f'delai.{table_name}'

    # Since string columns use the "object" dtype, pass in a (partial) schema
    # to ensure the correct BigQuery data type.
    write_mode = "WRITE_TRUNCATE" if is_first else "WRITE_APPEND"
    job_config = bigquery.LoadJobConfig(write_disposition=write_mode)
    
    job = client.load_table_from_dataframe(
        df, table_id, job_config=job_config
    )
    return job.result()
```

```python
def write_to_bq_in_chunks(df, table_name):
    index = 0
    chunk_size = 500000
    chunk = 0
    while index < len(df):
        print(f'Saving chunk', {chunk}, 'to bq.', 'Log:',{index}, {(chunk+1)*chunk_size})
        chunk_df = df[chunk*chunk_size: (chunk+1)*chunk_size]
        if chunk == 0:
            save_df_chunk_to_bq(chunk_df, table_name, is_first =True)

        else:
            save_df_chunk_to_bq(chunk_df, table_name, is_first =False)
        
        print(f'Chunk:', {chunk}, 'to bq.', 'Log:',{index}, {(chunk+1)*chunk_size}, 'DONE')
        chunk += 1
        index = chunk * chunk_size
```
